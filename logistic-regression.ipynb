# Logistic Regression

Predicts the probability for a given binary class.

**Inputs**
- $X \in [B, D_{in}]$
- $y \in [B, D_{out}]$, lets assume $D_{out} = 1$ with binary classification

**Model**

$z = X W + b$

$f(z) = \frac{1}{1+e^{-z}}$

**Loss**

Binary cross entropy

$\mathbb{l} = -\frac{1}{B} \sum_i (y_i \log f(z_i) + (1-y_i) \log (1-f(z_i)))$


**Gradient Derivation**

$$
\frac{d\mathbb{l}}{dW} = \frac{\partial l}{\partial f} \frac{\partial f}{\partial z} \frac{\partial z}{\partial W}
$$

$$
\frac{d\mathbb{l}}{db} = \frac{\partial l}{\partial f} \frac{\partial f}{\partial z} \frac{\partial z}{\partial b}
$$

Use the following derivatives:

$$\frac{\partial \mathbb{l}}{\partial f} = -\frac{1}{B}(\frac{y}{f} - \frac{1-y}{1-f})$$

$$\frac{\partial f}{\partial z} = -\frac{1}{(1+e^{-z})^2} \dot (-e^{-x}) \\
= \frac{e^{-x}}{(1+e^{-x})^2} \\
= f(1-f)
$$

$$
\frac{\partial z}{\partial W} = X
$$

$$
\frac{\partial z}{\partial b} = 1|_{(B \times 1)}
$$



we can finally reach:

$$
\frac{d\mathbb{l}}{dW} = \frac{1}{B} X^T(f-y)
$$

$$
\frac{d\mathbb{l}}{db} = \frac{1}{B}1|_{(1, B)}(f-y) = \frac{1}{B}\sum_{i}^{N}(f_i-y_i)
$$

## Data and Settings

```python
import torch 
import numpy as np
```

```python
B, Di, Do = 100, 10, 1

X = np.random.randn(B, Di)
y = np.random.randn(B, Do)
y = (y >= 0.5).astype(np.float32)

max_iters = 1000
lr = 1e-2
```

## Numpy Implementation

```python
def model(x, W, b):
    """
    Args:
        x: (B, Di)
        W: (Di, Do)
        b: (1,)
    Returns:
        (B, Do)
    """
    z = x @ W + b
    f = 1 / (1 + np.exp(-z))
    
    return f

def binary_cross_entropy(pred, target):
    """
    Args:
        pred: (B, Do)
        target: (B, Do)
    Returns:
        scalar
    """
    eps = 1e-8
    loss = - (target * np.log(pred + eps) + (1 - target) * np.log(1 - pred + eps))
    return np.mean(loss)

def gradients(pred, target, x):
    """
    Args:
        pred: (B, Do)
        target: (B, Do)
        x: (B, Di) 
    """
    B, _= pred.shape
    dw = 1/B * x.T @ (pred - target)
    db = 1/B * np.sum(pred - target)
    return dw, db
```

```python
# TRAINING 
# initialize weights
W = np.random.randn(Di, Do)
b = np.random.randn(1)

for i in range(max_iters):
    # forward 
    pred = model(X, W, b)

    # loss 
    loss = binary_cross_entropy(pred, y)

    # backward for gradient computation
    dw, db = gradients(pred, y, X)

    # update weights
    W -= lr * dw
    b -= lr * db

    # log 
    if i % 100 == 0:
        print(f"iter {i}: loss {loss:.4f}")
```
