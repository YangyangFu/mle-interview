# Gaussian Mixture Models

The main purpose of Gaussian mixture model is to estimate the density of a given data set. Gaussian mixture model assumes all the data points are generated from a mixture of a finite number of Gaussian distributions with unknown parameters. One can think of mixture models as generalizing k-means clustering to incorporate information about the covariance structure of the data as well as the centers of the latent Gaussians.


**Inputs**:
- $Y \in \mathbb{R}^{N, D}$: The labels for each sample, where $N$ is the number of samples.

**Outputs**:
- $\pi_k$: The mixing coefficients for each Gaussian component, where $k$ is the number of components.
- $\mu_k$: The mean vectors for each Gaussian component.
- $\Sigma_k$: The covariance matrices for each Gaussian component.



## Training
The model assumes that the data is generated from a mixture of $K$ Gaussian distributions, where each distribution is defined by its mean $\mu_k$ and covariance $\Sigma_k$. The mixing coefficients $\pi_k$ represent the proportion of data points that belong to each Gaussian component. The probability density function of the mixture model can be expressed as:

$$p(x) = \sum_{k=1}^{K} \pi_k \mathcal{N}(x | \mu_k, \Sigma_k)$$

Thus, the likelihood of the ground truth labels given the input data can be expressed as:
$$p(Y | X, \pi, \mu, \Sigma) = \prod_{n=1}^{N} \sum_{k=1}^{K} \pi_k \mathcal{N}(y_n | \mu_k, \Sigma_k)$$


Maximizing likelihood is to minimize the negative log-likelihood (NLL):

$$\mathcal{L}(\pi, \mu, \Sigma) = -\sum_{n=1}^{N} \log \left( \sum_{k=1}^{K} \pi_k \mathcal{N}(y_n | \mu_k, \Sigma_k) \right)$$



Expanding the Gaussian, we have:

$$\mathcal{L}(\pi, \mu, \Sigma) = -\sum_{n=1}^{N} \log \left( \sum_{k=1}^{K} \pi_k \frac{1}{(2\pi)^{D/2} |\Sigma_k|^{1/2}} \exp\left(-\frac{1}{2}(y_n - \mu_k)^T \Sigma_k^{-1} (y_n - \mu_k)\right) \right)$$

## Inference

The GMM models the probability distribution of the data, so we can use it to compute the probability of a new data point belonging to each Gaussian component. 
The GMM models donot directly predicts the label, but predicts the probability of a data point belonging to each Gaussian component. Thus it represents uncertainty in the prediction. 
The final prediction is made under such uncertainty, which is different from the deterministic prediction of k-means clustering.


### Most Likely Mode 
The most common way to make a prediction is to assign the data point to the Gaussian component with the highest posterior probability. This is equivalent to finding the most likely mode of the distribution. 

$$
\hat{k} = \arg\max_k \pi_k 
$$ 

$$
\hat{y} = \mu_{\hat{k}}
$$




### Minimum Expected Cost

Define a cost function $C(\hat{y})$ that measures the cost of predicting $\hat{y}$ in terms of downstream objectives, e.g., 
- distance to goal
- collision risk
- lane deviation

Then the predicted value $\hat{y}$ is the one that minimizes the expected cost:
$$
\hat{y} = \arg\min_{\hat{y}} \sum_{k=1}^{K} \pi_k C(\hat{y}, \mu_k) \sim \arg \min_k C(\mu_k)
$$

### Likelihood + Cost
The most practical way is to commbine the above both approaches, i.e., to find the mode with the lowest expected cost:

$$
\hat{k} = \arg\max_k (\log \pi_k - \lambda C(\mu_k))
$$

This will balance model confidence and task-specific objective.

### Stochastic Sampling
Another approach is to sample from the Gaussian mixture model to generate multiple possible predictions. This can be useful for tasks that require a diverse set of predictions, such as trajectory prediction in autonomous driving. The sampling process can be done by first sampling a component from the categorical distribution defined by $\pi_k$, and then sampling from the corresponding Gaussian distribution defined by $\mu_k$ and $\Sigma_k$.

Used for:
- uncertainty estimation
- generating diverse predictions
- simulation
